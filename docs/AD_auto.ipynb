```python
def edit_submit(submit_filename,Npcs_AD,N_cores,ad_dir_target):
    
    fad=open(AD_inputs_dir+submit_filename,'r')
    lines=fad.readlines()
    fad.close()

    lines[3]='target='+target+' \n'
    lines[4]='N='+str(Nmols)+' \n'
    lines[5]='n_submits='+ str(Npcs_AD) +' \n' 
    lines[6]='n_cores='+str(N_cores) +' \n'

    r=np.linspace(0,Nmols,Npcs_AD+1,dtype='int')
    for i in range(len(r)-1):   
        j0=r[i]+1
        jf=r[i+1]

        lines[7]='j0='+str(j0) +' \n'
        lines[8]='jf='+str(jf) +' \n'

        fad=open(ad_dir_target+'/'+submit_filename.split('.')[0]+'_'+str(i+1)+'.sh','w')
        for line in lines:
            fad.write(line)
        fad.close()
        
    return
```

```python
# Npcs_AD=1 # available PCs for AD
# N_cores=8 # number of max cores per PC

submit_filenames=['submit_ADdpfs.sh','submit_AD.sh','submit_ADresults.sh']

for submit_filename in submit_filenames:
    edit_submit(submit_filename,Npcs_AD,N_cores,ad_dir_target)
```

```python
def edit_submit_dpf(submit_filename,N_submits,N_cores):
    
    fad=open(AD_inputs_dir+'/CCAD_submits/'+submit_filename,'r')
    lines=fad.readlines()
    fad.close()

    r=np.linspace(0,Nmols,N_submits+1,dtype='int')
    for i in range(len(r)-1):   
        j0=r[i]+1
        jf=r[i+1]

        lines[19]='for ((I='+str(j0)+'; I<='+str(jf)+'; I++));do\n'

        fad=open(cwd+'/PRC_inputs/'+target+'/AD_inputs/CCAD_submits/'+submit_filename.split('.')[0]+'_'+str(i+1)+'.sh','w')
        for line in lines:
            fad.write(line)
        fad.close()
        
    return

def edit_submit_ad(submit_filename,N_submits,N_cores):
    
    fad=open(AD_inputs_dir+'/CCAD_submits/'+submit_filename,'r')
    lines=fad.readlines()
    fad.close()

    r=np.linspace(0,Nmols,N_submits+1,dtype='int')
    for i in range(len(r)-1):   
        j0=r[i]+1
        jf=r[i+1]

        lines[21]='for ((I='+str(j0)+'; I<='+str(jf)+'; I++));do\n'

        fad=open(cwd+'/PRC_inputs/'+target+'/AD_inputs/CCAD_submits/'+submit_filename.split('.')[0]+'_'+str(i+1)+'.sh','w')
        for line in lines:
            fad.write(line)
        fad.close()
        
    return

def edit_submit_res(submit_filename,N_submits,N_cores):
    
    fad=open(AD_inputs_dir+'/CCAD_submits/'+submit_filename,'r')
    lines=fad.readlines()
    fad.close()

    r=np.linspace(0,Nmols,N_submits+1,dtype='int')
    for i in range(len(r)-1):   
        j0=r[i]+1
        jf=r[i+1]

        lines[18]='for ((I='+str(j0)+'; I<='+str(jf)+'; I++));do\n'

        fad=open(cwd+'/PRC_inputs/'+target+'/AD_inputs/CCAD_submits/'+submit_filename.split('.')[0]+'_'+str(i+1)+'.sh','w')
        for line in lines:
            fad.write(line)
        fad.close()
        
    return
```
