```python
f=open(pwd+'/out_'+target+'_ad.sdf','r')
lines=f.readlines()
ix=[i for i in range(len(lines)) if lines[i]=='$$$$\n']
ix.insert(0,-1)
#ix.append(len(lines)-1)

Nmols=len(ix)-1
print('Number of molecules ',Nmols)
f.close()
```

```python
new_file = []

for i in range(len(ix) - 1):
    mol = lines[ix[i] + 1 : ix[i + 1] - 1]
    IX = np.arange(Nmols + 1)[1:][i]  # Obtener el Ã­ndice correspondiente

    # Verificar si IX estÃ¡ en dict_ix_scores antes de usarlo
    if IX not in dict_ix_scores:
        continue  # Si no estÃ¡, omitir esta iteraciÃ³n

    mol.append('M  END\n')
    mol.append('>  <IX> \n')
    mol.append(str(IX) + ' \n')
    mol.append('>  <Score_AD> \n')
    mol.append(str(dict_ix_scores[IX]) + ' \n')
    mol.append(' \n')
    mol.append('$$$$\n')

    for line in mol:
        new_file.append(line)
```
